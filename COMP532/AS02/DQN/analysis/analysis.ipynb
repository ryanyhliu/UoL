```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# 读取CSV文件并过滤以"Episode"开头的行
with open("data1.csv", "r", encoding="utf-8") as file:
    episode_lines = [line.strip() for line in file if line.startswith("Episode")]

# 读取过滤后的行作为DataFrame
df = pd.read_csv(StringIO("\n".join(episode_lines)), header=None, names=["Episode", "Time",  "Cumulative Reward"])

# 提取episode和累积奖励数据
episodes = df["Episode"].str.split("/").str[0].str.split(": ").str[1].astype(int)
rewards = df["Cumulative Reward"].str.split(": ").str[1].astype(float)



# 打印的表格大小
plt.figure(figsize=(16, 8))




plt.plot(episodes, rewards)

# 只保留最后500个episode的数据
# episodes_last_500 = episodes.tail(500)
# rewards_last_500 = rewards.tail(500)
# plt.plot(episodes_last_500, rewards_last_500)



plt.xlabel("Episode")
plt.ylabel("Cumulative Reward")
plt.grid(True)
plt.show()
```
